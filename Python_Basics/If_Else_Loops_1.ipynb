```python
num1 =int((input('Enter the first number')))
num2 =int((input('Enter the second number')))
op=input('Enter operation')
if op=="+":
    print(num1+num2)
elif op=="-":
    print(num1-num2)
elif op=="*":
    print(num1*num2)
elif op=="/":
    print(num1/num2)
if op=="%":
    print(num1%num2)
else :
    print('Enter valid operation')
```

# Nested Loops

```python
#example - unique pairs
for i in range(1,5):
    for j in range(1,5):
        print(i,j)
```

```python
rows=int(input('number of rows'))
for i in range (1,rows+1):
    for j in range(1,i+1):
        print("*",end='')
    print()
```

```python
rows=int(input('number of rows'))
for i in range (1,rows+1):
    for j in range (1,i+1):
        print(j,end="")
    print()
```

```python
rows=int(input('number of rows'))
for i in range (1,rows+1):
    for j in range (1,i+1):
        print(j,end="")
    for k in range(i-1,0,-1):
        print(k,end='')
    print()
```

```python
#break
for i in range(1,10):
    if i==5:
        break
    print(i)
```

```python
Lower = int(input('Ener lower range'))
Upper = int(input('Ener Upper range'))
for i in range(Lower,Upper+1):
    for j in range(2,i):
        if i%j==0:
            break
        else :
            print(i)
```
